```python
import pandas as pd
```

```python
df = pd.read_csv(
    "LnT_Helmet_Survey_Cleaned.csv"
)

df.head()
```

```python
df.shape
```

```python
def helemt_score(value):

    if value == "Always":
        return 3

    elif value == "Sometime":
        return 2

    elif value == "Never":
        return 1

    else:
        return 0
```

```python
df["Helemt_Compliance_Score"] = (
    df["Helemt_Frequency"]
    .apply(helemt_score)
)
```

```python
df[["Helemt_Frequency", "Helemt_Compliance_Score"]].head()
```

```python
df["Awareness_Score"] = 0
```

```python
df.loc[
df["ISI_Awareness"]=="Yes",
"Awareness_Score"
] += 1
```

```python
df.loc[
df["Helmet_Prevention_belief"]=="Yes",
"Awareness_Score"
] += 1
```

```python
df["Speed_Limit_Knowledge"].value_counts()
```

```python
df.loc[
df["Speed_Limit_Knowledge"]=="25 KM/H",
"Awareness_Score"
] += 1
```

```python
df[
[
"ISI_Awareness",
"Helmet_Prevention_belief",
"Awareness_Score"
]
].head()
```

```python
def risk_category(row):

    if (
        row["Helemt_Compliance_Score"] == 3
        and
        row["Awareness_Score"] >= 2
    ):
        return "Safety Champion"


    elif row["Helemt_Compliance_Score"] == 1:
        return "High Risk"


    else:
        return "Moderate Risk"
```

```python
df["Risk_Category"] = df.apply(
    risk_category,
    axis=1
)
```

```python
df["Risk_Category"].value_counts()
```

```python
def behaviour_gap(row):

    if (
        row["Awareness_Score"] >=2
        and
        row["Helemt_Compliance_Score"] <=1
    ):
        return "Awareness Without Action"

    else:
        return "Aligned Behaviour"
```

```python
df["Behaviour_Gap"] = df.apply(
    behaviour_gap,
    axis=1
)
```

```python
df["Behaviour_Gap"].value_counts()
```

```python
def age_segment(age):

    if age <=20:
        return "Young Rider"

    elif age <=30:
        return "Adult Rider"

    else:
        return "Experienced Rider"


df["Age_Segment"] = df["Age"].apply(age_segment)
```

```python
df.to_csv(
"LnT_Safety_Analytics_Final.csv",
index=False
)
```
